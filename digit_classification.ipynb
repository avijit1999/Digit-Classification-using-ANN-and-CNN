```python
import tensorflow as tf
```

```python
from tensorflow import keras
from tensorflow.keras import datasets,layers,models
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
plt.matshow(X_train[0])
```

```python
X_train_new=X_train.reshape(60000,28*28)
X_test_new=X_test.reshape(10000,28*28)
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits= True),
    metrics=['accuracy']
)

model.fit(X_train_new,y_train,epochs=5)
```

```python
plt.matshow(X_test[1])
```

```python
y_predict=model.predict(X_test_new)
y_predict[1]
```

```python
np.argmax(y_predict[1])
```

```python
y_predict_lebel=[np.argmax(i) for i in y_predict]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predict_lebel)
```

```python
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predict')
plt.ylabel("Truth")
```

```python
model=keras.Sequential([
    keras.layers.Dense(600,input_shape=(784,),activation='relu'),
    keras.layers.Dense(400,activation='relu'),
    keras.layers.Dense(200,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')

])

model.compile(
    optimizer='adam',
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits= True),
    metrics=['accuracy']
)

model.fit(X_train_new,y_train,epochs=10)
```

```python
y_predict=model.predict(X_test_new)
y_predict[1]
```

```python
y_predict_lebel=[np.argmax(i) for i in y_predict]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predict_lebel)
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predict')
plt.ylabel("Truth")
```

```python
X_train = X_train.reshape(-1,28,28,1)
X_train.shape
```

```python
X_test = X_test.reshape(-1,28,28,1)
X_test.shape
```

```python
model=keras.Sequential([
    #cnn
    layers.Conv2D(filters=30,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    #ann
    layers.Flatten(),
    layers.Dense(100,activation='relu'),
    layers.Dense(10,activation='sigmoid')

])

model.compile(
    optimizer='adam',
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits= True),
    metrics=['accuracy']
)

model.fit(X_train,y_train,epochs=5)
```

```python
y_predict=model.predict(X_test)
#y_predict[1]
y_predict_lebel=[np.argmax(i) for i in y_predict]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predict_lebel)
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predict')
plt.ylabel("Truth")
```

```python
model.evaluate(X_test,y_test)
```
